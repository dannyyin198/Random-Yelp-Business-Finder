```python
class Business:
    def __init__(self,business_id,name,address,city,state,postal_code,stars,review_count):
        self.business_id = business_id
        self.name = name
        self.address = address
        self.city = city
        self.state = state
        self.postal_code = postal_code
        self.stars = stars
        self.review_count = review_count


        
def get_star_business(businesses, rating):
    star_business = []
    for business in businesses:
        if business.stars >= rating:
            star_business.append(business)
    return star_business

def get_business_in_city(businesses, city):
    city_business = []
    for business in businesses:
        if business.city.lower() == city.lower():
            city_business.append(business)
    return city_business

def get_business_in_zipcode(businesses, zipcode):
    zipcode_business = []
    for business in businesses:
        if business.postal_code == zipcode:
            zipcode_business.append(business)
    return zipcode_business

columns = ['business_id', 'name', 'address', 'city', 'state', 'postal_code', 'stars', 'review_count']

dfbusiness = df[columns]

businesses = []

for index, row in dfbusiness.iterrows():
    business = Business(row['business_id'], row['name'], row['address'], str(row['city']), row['state'], row['postal_code'], row['stars'], row['review_count'])
    businesses.append(business)

rating = int(input("Enter the minimum rating of the business: "))
city = input("Enter the city of the business: ")
zipcode = input("Enter the zipcode of the business: ")

star_business = get_star_business(businesses, rating)
city_business = get_business_in_city(star_business, city)
zipcode_business = get_business_in_zipcode(city_business, zipcode)

random_business = random.sample(zipcode_business, 10)

print (f"10 RANDOM BUSINESSES WITH RATING >= {rating}, IN CITY: {city}, AND ZIPCODE: {zipcode}")
print ('----------')
for business in random_business[:10]:
    pprint(f"Name: {business.name}, Address: {business.address}, City: {business.city}, State: {business.state}, Postal code: {business.postal_code}, RATING: {business.stars}, Number of Reviews: {business.review_count}")
```
